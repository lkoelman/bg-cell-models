Numerical methods for solution of the Cable Equation

# Cable Equation

Cable equation as given in [Tsuyuki (2016)][Tsuyuki2016]:

$$ C_m \frac{\delta V(x,t)}{\delta t} = \frac{a}{2R_a} \frac{\delta^2 V(x,t)}{\delta x^2} - gV(x,t) + I_{ext} $$

where $a$ is the diameter (also denoted as $d$), $R_a$ is the axial resistivity, and $g$ is the membrane conductance.

# Discretized Cable Equation

## Spatial Discretization

By discretizing the cable equation in the spatial dimension and using the 2nd order central difference (in space) we get the following equation for compartment `j`:

$$ C_m \frac{\delta V_j(t)}{\delta t} + \sum_i I_{m,i} = \frac{a}{4R_a} \frac{[V_{j+1}(t)-V_j(t)] + [V_{j-1}(t) - V_j(t)]}{\Delta x^2} $$


Where the LHS are the membrane currents and the RHS the axial currents. This is a specific formulation of the cable equation, i.e. in terms of current density. Its units are $\frac{[mA]}{[cm^2]}$ if the following quantities are used ([Sterrat et al. (2011), table 2.3][GilliesWillshaw2011]):

## Temporal Discretization

To discretize this equation in the temporal dimension we can take the [backward difference in time][wp:finite_difference], i.e.:

$$ f'(x) = \lim_{h \to 0} \frac{f(x) - f(x-h)}{h} $$

This results in:

$$ -\sigma V_{j-1}(t+\Delta t) + (1+2\sigma+\gamma) V_j(t+\Delta t) - \sigma \sum_{k}^{child} V_{j+k}(t+\Delta t) = V_j(t) $$

where the contribution of axial currents from neighboring compartments is

$$ \sigma = \frac{a \Delta t}{2 R_a C_m (\Delta x)^2} $$

and the contribution of internal compartmental conductances is

$$ \gamma = \frac{g_i}{C_m} $$
